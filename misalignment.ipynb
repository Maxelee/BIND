```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

def plot_1P_param_family(param_tag, P_params, shapes, value_col, cmap_name="plasma"):
    """
    param_tag : str
        Parameter index like 'p1', 'p23', 'p35', ...
        Will look for sims named:
            1P_<param_tag>_n2, 1P_<param_tag>_n1, 1P_<param_tag>_1, 1P_<param_tag>_2
    P_params  : pd.DataFrame
        Parameter table with '#Name' and a column value_col to color by.
    shapes    : dict-like
        Must contain 'sim_id' and fields 'truth_*' / 'gen_*'.
    value_col : str or int
        Column name (preferred) or integer index in P_params to use for coloring.
    cmap_name : str
        Matplotlib colormap name.
    """

    # 1. Build sim names for this parameter
    names = [f"1P_{param_tag}_n2",
             f"1P_{param_tag}_n1",
             f"1P_{param_tag}_1",
             f"1P_{param_tag}_2"]

    # 2. Extract the parameter values for these runs, preserving order
    sub = (
        P_params[P_params["#Name"].isin(names)]
        .set_index("#Name")
        .loc[names]  # enforce n2, n1, 1, 2 order
    )

    if isinstance(value_col, int):
        vals = sub.iloc[:, value_col].to_numpy()
    else:
        vals = sub[value_col].to_numpy()

    vals = np.asarray(vals, dtype=float)

    # Check for degenerate case: all same value
    vmin, vmax = vals.min(), vals.max()
    if vmin == vmax:
        # expand a bit to avoid all identical colors in Normalize
        vmin -= 1e-6
        vmax += 1e-6

    norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)
    cmap = mpl.cm.get_cmap(cmap_name)
    colors = cmap(norm(vals))  # RGBA color per sim

    # 3. Build masks over halos for each sim
    sim_ids = np.asarray(shapes["sim_id"])
    masks = [sim_ids == name for name in names]

    components = ["DM", "Gas", "Stars"]
    rows = [
        ("q",  "a0"),
        ("q",  "a1"),
        ("e1", "a0"),
        ("e1", "a1"),
        ("e2", "a0"),
        ("e2", "a1"),
    ]

    fig, axes = plt.subplots(len(rows), len(components), figsize=(12, 20), squeeze=False)

    for i, (stat, a_idx) in enumerate(rows):
        for j, comp in enumerate(components):
            ax = axes[i, j]

            truth_key = f"truth_{comp}_{stat}_{a_idx}"
            gen_key   = f"gen_{comp}_{stat}_{a_idx}"

            # plot all halos for each sim in this family
            for name, mask, color in zip(names, masks, colors):
                truth = np.asarray(shapes[truth_key][mask])
                gen   = np.asarray(shapes[gen_key][mask])

                ax.scatter(
                    truth,
                    gen,
                    s=4,
                    color=color,       # fixed color per sim
                    alpha=0.4,
                    rasterized=True,
                    label=name if (i == 0 and j == 0) else None,
                )

            # Limits / 1:1 line from all halos in family
            combined_mask = np.logical_or.reduce(masks)
            all_truth = np.asarray(shapes[truth_key][combined_mask])
            all_gen   = np.asarray(shapes[gen_key][combined_mask])

            lo = np.nanmin([all_truth.min(), all_gen.min()])
            hi = np.nanmax([all_truth.max(), all_gen.max()])

            ax.plot([lo, hi], [lo, hi], c="k", lw=1, ls="--")
            ax.set_xlim(lo, hi)
            ax.set_ylim(lo, hi)

            if i == 0:
                ax.set_title(comp, fontsize=14)
            if j == 0:
                ax.set_ylabel(f"{stat} {a_idx}\nGenerated")
            if i == len(rows) - 1:
                ax.set_xlabel("Truth")

    # colorbar for the chosen parameter
    mappable = mpl.cm.ScalarMappable(norm=norm, cmap=cmap)
    cbar = fig.colorbar(mappable, ax=axes, shrink=0.95, pad=0.02)
    cbar.set_label(str(value_col))

    # legend for which sim name = which color
    axes[0, 0].legend(
        loc="lower right",
        fontsize=8,
        frameon=False,
        title=fr"$1P\_{param_tag}$ runs",
    )

    plt.tight_layout()
    plt.show()
```

```python
# Color by Omega0 column
plot_1P_param_family("p1",  P_params, shapes, value_col="Omega0")

# Color by, say, sigma8 if that’s a column name
plot_1P_param_family("p23", P_params, shapes, value_col="sigma8")

# Or by numerical column index (not recommended, but possible)
plot_1P_param_family("p35", P_params, shapes, value_col=1)
```
