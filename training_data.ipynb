```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
halo = 0
train_data_path = '/mnt/home/mlee1/ceph/train_data_rotated2_128_cpu/train/sim_0'
halo_rots = np.arange(10)
halo_paths = [f'{train_data_path}/sim_0_halo_{halo}_rot_{rot}.npz' for rot in halo_rots]
```

```python
halos = [np.load(h) for h in halo_paths]
```

```python
# conditions 
fig, axs = plt.subplots(nrows = 2, ncols=4, figsize=(12, 6), sharex='col', sharey='col')
extents = np.array([-6.25/2, 6.25/2, -6.25/2, 6.25/2])
for row in range(2):
    condition = np.log10(halos[row]['condition'])
    large_scale = np.log10(halos[row]['large_scale']   )
    axs[row, 0].imshow(condition[:, :], cmap='viridis', vmin=8, vmax=12, extent=extents)
    for i in range(1, 4):
        axs[row, i].imshow(large_scale[i-1, :, :], cmap='viridis', vmin=8, vmax=12, extent=2**i*extents)
```
